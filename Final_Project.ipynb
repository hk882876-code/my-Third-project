```python
# Simple Calculator Program

# Ask how many calculations the user wants to perform
n = int(input("How many calculations do you want to perform? "))

for i in range(n):
    print("\nCalculation", i + 1)

    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))


    operator = input("Enter operator (+, -, *, /): ")
    if operator == '+':
        result = num1 + num2
        print("Result:", result)

    elif operator == '-':
        result = num1 - num2
        print("Result:", result)

    elif operator == '*':
        result = num1 * num2
        print("Result:", result)

    elif operator == '/':
        if num2 == 0:
            print("Error: Division by zero is not allowed.")
        else:
            result = num1 / num2
            print("Result:", result)

    else:
        print("Error: Invalid operator.")
```
